```
import yfinance as yf
import pandas as pd
from hdfs3 import HDFileSystem
from mrjob.job import MRJob
import re
import sys
import csv
```

```
class TopStocks(MRJob):

   def mapper(self, key, value):
      entries = csv.reader([value]).__next__()
      if int(entries[0][5] + entries[0][1]) == 10 or int(entries[0][5] + entries[0][1]) == 11 or int(entries[0][5] + entries[0][1]) == 12:    
         apple_stock = float(entries[1])
         ibm_stock = float(entries[2])
         yield('APPL', apple_stock)
         yield('IBM', ibm_stock)

   def reducer(self, key, values):
      top_ten_stocks = []

      for stock in values:
         top_ten_stocks.append(stock)
         top_ten_stocks.sort(reverse = True)
         top_ten_stocks = top_ten_salaries[:10]

      yield(key, top_ten_salaries[0])
```
